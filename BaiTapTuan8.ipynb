```python
#W7A1
def binary_search(a, l, r, i):
    while(l<=r):
        m = (l+r)//2
        if (a[m] == i):
            return m
        elif (a[m] > i): 
            r = m-1
        else: l = m + 1
    return -1
a = list(map(int, input().split()))
i = int(input())
print(binary_search(a, 0, len(a)-1, i))
```

```python
#W7A2
def binary_search_l(a, l, r, i):
    res = -1
    while(l<=r):
        m = (l+r)//2
        if (a[m] >= i):
            r = m-1
            res = m
        else: l = m+1
    return res
def binary_search_r(a, l, r, i):
    res = -1
    while(l<=r):
        m = (l+r)//2
        if (a[m] <= i):
            l = m+1
            res = m
        else: r = m-1
    return res
a = list(map(int, input().split()))
a.sort()
i = int(input())
lef = binary_search_l(a, 0, len(a)-1, i)
rig = binary_search_r(a, 0, len(a)-1, i)
if (rig == -1): print(0)
else: print(rig - lef + 1)
```

```python
#W7A3
# bubble sort
def bubble_sort(a):
    for i in range(len(a)):
        check = True
        for j in range(len(a)-1):
            if (a[j] > a[j+1]): 
                a[j], a[j+1] = a[j+1], a[j]
                check = False
        if (check): break
    return a
a = list(map(int, input().split()))
# selection_sort
def selection_sort(a):
    for i in range(len(a)):
        mn = a[i]
        for j in range(i, len(a)):
            mn = min(mn, a[i])
        a[i] = mn
    return a
print(bubble_sort(a))
print(selection_sort(a))
```

```python
#W7A4
m = {}
def solve(a):
    a.sort()
    tmp = a[0]
    dem = 1
    res = 0
    for i in range(1, len(a)):
        if (a[i] != tmp):
            m[tmp] = dem
            res = max(res, dem)
            tmp = a[i]
            dem = 1
        else: dem += 1
    m[a[len(a)-1]] = dem
    return res
a = list(map(int, input().split()))
mx = solve(a)
for i in a:
    if (m[i] == mx):
        print(f"{i} xuat hien nhieu nhat, som nhat,{m[i]} lan")
        break
```
