```python
#W6A9
m1 = {}
nhap1 = map(str, input().split())
for i in nhap1:
    a, b = map(str, i.split(':'))
    if (a not in m1): m1[a] = int(b)
    else: m1[a] += int(b)
nhap2 = map(str, input().split())
for i in nhap2:
    a, b = map(str, i.split(':'))
    if (a not in m1): m1[a] = int(b)
    else: m1[a]+=int(b)
m2 = sorted(m1)
for i in m2: print(i, m1[i])
```

```python
#W6A10
lst = list(map(int, input().split()))
k = int(input())
res = []
ln = len(lst)
for i in range(ln-1):
    for j in range(i+1, ln):
        if (lst[i] + lst[j] == k): res.append((lst[i],lst[j]))
res = sorted(res, key = lambda x: x[1], reverse = True)
print(res)
```
