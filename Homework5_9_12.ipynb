```python
#W6A21
m = {}
lst = list(map(str, input().split()))
for i in range(len(lst)):
    if (lst[i] not in m):
        m[lst[i]] = [i]
    else: m[lst[i]].append(i)
for i in m:
    m[i] = tuple(m[i])
print(m)
```

```python
#W6A23
def rm(a, x):
    res = []
    for i in a:
        if (i >= x): res.append(i)
    return res
lst = list(map(int, input().split()))
x = int(input())
print(rm(lst, x))
```

```python
#W6A24
def xuly(a):
    khong = []
    khac = []
    for i in a:
        if (i == 0): khong.append(i)
        else: khac.append(i)
    return (khong+khac, khac+khong)
lst = list(map(int, input().split()))
a, b = xuly(lst)
print(a)
print(b)
```

```python
#W6A26
lst = list(map(int, input().split()))
res  = 0
i = 1
while(i < len(lst)):
    if (lst[i] < lst[i-1]): 
        res += lst[i-1]
        while (lst[i] < lst[i-1] and i < len(lst)): 
            i+=1
    i+=1
print(res)
```

```python
#W6A27
def xuly(a, x):
    res = []
    for i in range(len(a)):
        if (a[i] < x): res.append(i)
    return res
lst = list(map(int, input().split()))
x = int(input())
res = xuly(lst,x)
if (len(res) == 0): print(-1)
else: print(res)
```
