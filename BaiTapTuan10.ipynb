```python
def merge(a, b):
    c = [0]*(len(a) + len(b))
    dem = 0
    i, j = 0, 0
    while(i < len(a) and j < len(b)):
        if (a[i] > b[j]):
            c[dem] = b[j]
            j += 1
            dem += 1
        else: 
            c[dem] = a[i]
            i += 1
            dem += 1
    for i in range(i, len(a)): 
        c[dem] = a[i]
        dem += 1
    for i in range(j, len(b)): 
        c[dem] = b[j]
        dem += 1
    return c
tmp, res = [], []
n, m = map(int, input().split())
for i in range(n):
    if (i != 0):
        res = list(map(int, input().split()))
        tmp = merge(res, tmp)
    else:
        tmp = list(map(int, input().split()))
k = int(input())
print(tmp[k-1])
```

```python
def Median(nums):
    a, b, c, d, e = nums

    # 1–2: đảm bảo a <= b
    if a > b:
        a, b = b, a

    # 3–4: đảm bảo c <= d
    if c > d:
        c, d = d, c

    # 5: loại min toàn cục
    if a > c:
        a, c = c, a
        b, d = d, b

    # lúc này: a là nhỏ nhất → bỏ a

    # 6: so sánh b, c, e để lấy trung vị
    if b > c:
        b, c = c, b

    # trung vị nằm giữa b, c, e
    if e < b:
        return b
    elif e > c:
        return c
    else:
        return e
num = list(map(int, input().split()))
Median(num)
```
