```python
#W5A3
def check(n):
    for i in range(2, int(n**0.5)+1):
        if (n%i == 0): return False
    return n > 1
n = int(input())
if (check(n)):
    print("So nguyen to")
else: print("Khong la so nguyen to")
```

```python
#W5A4
def solve(n):
    sum = 0
    for i in range(1, int(n**0.5)+1):
        if (n%i == 0):
            sum += int(n/i)
            if (i*i != n): sum += i
    return sum
n = int(input())
if (2*n == solve(n)):
    print("Day la so hoan hao")
else: print("Day khong phai la so hoan hao")
```

```python
#W5A6
def solve(n):
    res = 1
    for i in range(1, n+1):
        res *= i
    return res
n = int(input())
print(solve(n))
```

```python
#W5A7
def solve(a, b):
    tru = a - b
    cong = a + b
    chia = round(a/b,2)
    nhan = a*b
    return tru, cong, chia, nhan
a, b = map(int, input().split())
print(solve(a, b))
```

```python
#W5A8
def solve(a, b):
    dem = 0
    if (len(a) > len(b)): a, b = b, a
    for i in range(len(b)):
        if (i < len(a)):
            if (a[i] != b[i]): dem += 1
        else: dem += 1
    return dem
a, b = map(str, input().split())
print(solve(a, b))
```

```python
#W5A9
def sum(n):
    tong = 0
    while(n > 0):
        tong += n%10
        n //= 10
    return tong
n = int(input())
print(sum(n))
```

```python
#W5A10
def check(a, b):
    m1 = {}    
    s1 = set()
    for i in range(0, len(a)):
        if (a[i] in m1):
            if (m1[a[i]] != b[i]): return False
        else:
            if (b[i] in s1): return False
            m1[a[i]] = b[i]
            s1.add(b[i])
    return True

a, b = map(str, input().split())
if (check(a, b)):
    print("dang cau")
else: print("khong phai dang cau")
```
