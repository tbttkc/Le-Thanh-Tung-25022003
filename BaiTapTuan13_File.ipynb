```python
#W11A4
from collections import Counter
import string

def solve(path1, path2):
    def normalize(s):
        return ''.join(c.lower() for c in s if c.isalpha())
    with open(path1, 'r') as f1, open(path2, 'r') as f2:
        s1 = normalize(f1.read())
        s2 = normalize(f2.read())
    return Counter(s1) == Counter(s2)
print(solve('a.txt', 'b.txt'))
```

```python
#W11A10
def get(path):
    with open(path, 'r') as f:
        return list(map(int, f.read().split()))


def rob(hourses]):
    if not hourses:
        return 0
    if len(hourses) == 1:
        return hourses[0]

    dp1, dp2 = 0, 0
    for money in hourses:
        dp1, dp2 = dp2, max(dp2, dp1 + money)
    return dp2
```

```python
#W11A12
def solve(path):
    with open(path, 'r') as f:
        n = int(f.readline().strip())
        nums = list(map(int, f.readline().split()))
    nums.sort()
    return max(nums[-1] * nums[-2] * nums[-3],
               nums[0] * nums[1] * nums[-1])
print(solve('a.txt'))
```

```python
#W11A13
def solve(path):
    with open(path, 'r') as f:
        n = int(f.readline().strip())
        nums = list(map(int, f.readline().split()))
    result = []
    zero_count = 0
    for x in nums:
        if x == 0:
            zero_count += 1
        else:
            result.append(x)
    result.extend([0] * zero_count)
    return result
print(solve('a.txt'))
```

```python
#W11A14
def solve(path):
    with open(path, 'r') as f:
        lines = f.read().splitlines()
    times = []
    for line in lines[1:]:
        parts = line.split(',')
        times.append(float(parts[1]))
    avg = sum(times) / len(times)
    return round(avg, 2)
print(solve('a.txt'))
```

```python
#W11A15
def solve(path):
    result = []
    with open(path, 'r', encoding='utf-8') as f:
        lines = f.read().splitlines()
    for line in lines[1:]:
        parts = [p.strip() for p in line.split(',')]
        movie = parts[1]
        description = parts[2].lower()
        rating = float(parts[3])
        if rating > 8.0 and description != "boring":
            result.append(movie)
    if not result:
        return ["All are boring!"]
    return result
print(solve('a.txt'))
```

```python
#W11A16
import math
def solve(path):
    with open(path, 'r') as f:
        lines = [line.strip() for line in f if line.strip() != ""]
    i = 0
    while i < len(lines):
        shape_type = lines[i].split(":")[1].strip()
        params = lines[i + 1]
        i += 2
        if shape_type == "SQUARE":
            side = float(params.split(":")[1])
            perimeter = 4 * side
        elif shape_type == "RECTANGLE":
            parts = params.split()
            width = float(parts[1])
            height = float(parts[3])
            perimeter = 2 * (width + height)
        elif shape_type == "CIRCLE":
            radius = float(params.split(":")[1])
            perimeter = 2 * math.pi * radius
        print(f"{perimeter:.2f}")
print(solve('a.txt'))
```

```python
#W11A17
def solve(path):
    with open(path, 'r') as f:
        nums = list(map(int, f.read().split()))
    n = len(nums)
    result = [1] * n
    prefix = 1
    for i in range(n):
        result[i] = prefix
        prefix *= nums[i]
    suffix = 1
    for i in range(n - 1, -1, -1):
        result[i] *= suffix
        suffix *= nums[i]
    return result
print(solve('a.txt'))
```

```python
#W11A18
import math
def solve(path):
    with open(path, 'r') as f:
        lines = f.read().splitlines()
    src_parts = lines[0].split()
    des_parts = lines[1].split()
    vel_parts = lines[2].split()
    x1 = float(src_parts[2])
    y1 = float(src_parts[4])
    x2 = float(des_parts[2])
    y2 = float(des_parts[4])
    v = float(vel_parts[1])
    distance = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
    return round(distance / v, 2)
print(solve('a.txt'))
```

```python
#W11A20
def solve(path):
    with open(path, 'r') as f:
        lines = f.read().splitlines()
    M = int(lines[0])
    img = [list(map(int, lines[i + 1].split())) for i in range(M)]
    res = [[0] * M for _ in range(M)]
    for i in range(M):
        for j in range(M):
            total = 0
            count = 0
            for dx in [-1, 0, 1]:
                for dy in [-1, 0, 1]:
                    ni, nj = i + dx, j + dy
                    if 0 <= ni < M and 0 <= nj < M:
                        total += img[ni][nj]
                        count += 1
            res[i][j] = total // count
    for row in res:
        print(*row)
solve('a.txt')
```
